```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Baca data
df = pd.read_csv(ratings_path)
df = pd.DataFrame(df)

# Periksa data
# print("Jumlah data:", len(df))
# print("\nStruktur data:")
# print(df.head())
# print("\nInformasi data:")
# print(df.info())
# print("\nStatistik deskriptif:")
# print(df.describe())

class SVDMatrixFactorization:
    def __init__(self, ratings, n_factors, learning_rate, regularization, n_epochs):
        self.ratings = ratings
        self.n_users = ratings['userId'].nunique()
        self.n_items = ratings['movieId'].nunique()
        self.n_factors = n_factors
        self.learning_rate = learning_rate
        self.regularization = regularization
        self.n_epochs = n_epochs

        # Mapping untuk user_id dan movie_id
        self.user_mapping = {user_id: idx for idx, user_id in enumerate(ratings['userId'].unique())}
        self.item_mapping = {movie_id: idx for idx, movie_id in enumerate(ratings['movieId'].unique())}
        self.reverse_user_mapping = {idx: user_id for user_id, idx in self.user_mapping.items()}
        self.reverse_item_mapping = {idx: movie_id for movie_id, idx in self.item_mapping.items()}

        # Inisialisasi matriks latent factor
        np.random.seed(42)
        self.user_factors = np.random.normal(scale=0.1, size=(self.n_users, self.n_factors))
        self.item_factors = np.random.normal(scale=0.1, size=(self.n_items, self.n_factors))

        # Global bias, user bias, item bias
        self.global_mean = np.mean(ratings['rating'])
        self.user_bias = np.zeros(self.n_users)
        self.item_bias = np.zeros(self.n_items)

        # Log untuk menyimpan error
        self.train_rmse_log = []
        self.test_rmse_log = []

    def get_user_idx(self, user_id):
        return self.user_mapping.get(user_id)

    def get_item_idx(self, movie_id):
        return self.item_mapping.get(movie_id)

    def get_original_user_id(self, user_idx):
        return self.reverse_user_mapping.get(user_idx)

    def get_original_movie_id(self, item_idx):
        return self.reverse_item_mapping.get(item_idx)

    def train(self, train_data, test_data=None):
        """
        Melatih model SVD dengan SGD (Stochastic Gradient Descent)
        """
        print("Mulai pelatihan model...")

        for epoch in range(self.n_epochs):
            # Shuffle data
            train_data = train_data.sample(frac=1).reset_index(drop=True)

            # Hitung error
            errors = []

            # Update parameter untuk setiap rating
            for index, row in train_data.iterrows():
                user_id, movie_id, rating = row['userId'], row['movieId'], row['rating']

                user_idx = self.get_user_idx(user_id)
                item_idx = self.get_item_idx(movie_id)

                if user_idx is None or item_idx is None:
                    continue

                # Hitung prediksi
                prediction = self.predict(user_id, movie_id)

                # Hitung error
                error = rating - prediction
                errors.append(error ** 2)

                # Update bias
                self.user_bias[user_idx] += self.learning_rate * (error - self.regularization * self.user_bias[user_idx])
                self.item_bias[item_idx] += self.learning_rate * (error - self.regularization * self.item_bias[item_idx])

                # Update latent factors
                user_factors_prev = self.user_factors[user_idx].copy()
                item_factors_prev = self.item_factors[item_idx].copy()

                self.user_factors[user_idx] += self.learning_rate * (error * item_factors_prev - self.regularization * user_factors_prev)
                self.item_factors[item_idx] += self.learning_rate * (error * user_factors_prev - self.regularization * item_factors_prev)

            # Hitung RMSE untuk data pelatihan
            train_rmse = np.sqrt(np.mean(errors))
            self.train_rmse_log.append(train_rmse)

            # Evaluasi pada data uji jika disediakan
            test_rmse = None
            if test_data is not None:
                test_predictions = []
                test_true_ratings = []

                for index, row in test_data.iterrows():
                    user_id, movie_id, rating = row['userId'], row['movieId'], row['rating']
                    if self.get_user_idx(user_id) is not None and self.get_item_idx(movie_id) is not None:
                        pred = self.predict(user_id, movie_id)
                        test_predictions.append(pred)
                        test_true_ratings.append(rating)

                test_rmse = np.sqrt(mean_squared_error(test_true_ratings, test_predictions))
                self.test_rmse_log.append(test_rmse)

            # Tampilkan kemajuan pelatihan tiap 5 epoch
            if (epoch + 1) % 5 == 0:
                if test_rmse is not None:
                    print(f"Epoch {epoch+1}/{self.n_epochs} - Train RMSE: {train_rmse:.4f}, Test RMSE: {test_rmse:.4f}")
                else:
                    print(f"Epoch {epoch+1}/{self.n_epochs} - Train RMSE: {train_rmse:.4f}")

        print("Pelatihan selesai!")
        return self.train_rmse_log, self.test_rmse_log

    def predict(self, user_id, movie_id):
        """
        Memprediksi rating untuk pasangan user-movie
        """
        user_idx = self.get_user_idx(user_id)
        item_idx = self.get_item_idx(movie_id)

        # Jika user atau movie tidak dikenal, kembalikan rating rata-rata global
        if user_idx is None or item_idx is None:
            return self.global_mean

        # Prediksi = global_mean + user_bias + item_bias + dot_product(user_factors, item_factors)
        pred = self.global_mean + self.user_bias[user_idx] + self.item_bias[item_idx] + \
               np.dot(self.user_factors[user_idx], self.item_factors[item_idx])

        # Batasi prediksi dalam rentang rating yang valid (1-5)
        return max(1, min(5, pred))

    def plot_learning_curve(self):
        """
        Plot learning curve untuk melihat performa model
        """
        plt.figure(figsize=(10, 6))
        plt.plot(range(1, self.n_epochs + 1), self.train_rmse_log, label='Training RMSE')
        if self.test_rmse_log:
            plt.plot(range(1, self.n_epochs + 1), self.test_rmse_log, label='Test RMSE')
        plt.xlabel('Epoch')
        plt.ylabel('RMSE')
        plt.title('Learning Curve')
        plt.legend()
        plt.grid(True)
        plt.show()

    def get_similar_movies(self, movie_id, top_n=10):
        """
        Menemukan film yang mirip berdasarkan latent factors
        """
        item_idx = self.get_item_idx(movie_id)

        if item_idx is None:
            print(f"Film dengan ID {movie_id} tidak ditemukan.")
            return []

        # Hitung cosine similarity antara film yang dicari dan semua film lainnya
        target_factors = self.item_factors[item_idx]
        similarities = []

        for idx, factors in enumerate(self.item_factors):
            if idx != item_idx:
                similarity = np.dot(target_factors, factors) / (np.linalg.norm(target_factors) * np.linalg.norm(factors))
                similarities.append((self.get_original_movie_id(idx), similarity))

        # Urutkan berdasarkan similaritas dan kembalikan top_n
        return sorted(similarities, key=lambda x: x[1], reverse=True)[:top_n]

    def recommend_movies(self, user_id, top_n=10, exclude_rated=True):
        """
        Rekomendasikan film untuk user tertentu
        """
        user_idx = self.get_user_idx(user_id)

        if user_idx is None:
            print(f"User dengan ID {user_id} tidak ditemukan.")
            return []

        # Dapatkan film yang sudah dinilai oleh user jika exclude_rated=True
        rated_movies = set()
        if exclude_rated:
            rated_movies = set(self.ratings[self.ratings['userId'] == user_id]['movieId'])

        predictions = []
        # Prediksi rating untuk semua film
        for movie_idx in range(self.n_items):
            movie_id = self.get_original_movie_id(movie_idx)

            # Lewati film yang sudah dinilai jika exclude_rated=True
            if exclude_rated and movie_id in rated_movies:
                continue

            pred_rating = self.predict(user_id, movie_id)
            predictions.append((movie_id, pred_rating))

        return sorted(predictions, key=lambda x: x[1], reverse=True)[:top_n]

    def calculate_intra_list_diversity(self, recommendations):
        """
        Menghitung keragaman (diversity) rata-rata antar film dalam daftar rekomendasi
        berdasarkan cosine similarity latent factors.
        """
        if len(recommendations) <= 1:
            return 0

        total_pairs = 0
        similarity_sum = 0

        for i in range(len(recommendations)):
            movie_id_1 = recommendations[i][0]
            idx1 = self.get_item_idx(movie_id_1)
            for j in range(i+1, len(recommendations)):
                movie_id_2 = recommendations[j][0]
                idx2 = self.get_item_idx(movie_id_2)
                sim = np.dot(self.item_factors[idx1], self.item_factors[idx2]) / (
                    np.linalg.norm(self.item_factors[idx1]) * np.linalg.norm(self.item_factors[idx2])
                )
                similarity_sum += sim
                total_pairs += 1

        average_similarity = similarity_sum / total_pairs
        diversity = 1 - average_similarity  # Semakin tinggi diversity, semakin rendah similarity rata-rata
        return diversity

    def evaluate_recommendation_quality(self, user_id, num_recommendations=10):
        """
        Evaluasi rekomendasi untuk seorang user berdasarkan diversity dan novelty.
        """
        recommendations = self.recommend_movies(user_id, top_n=num_recommendations)

        # Hitung popularitas film (misalnya jumlah rating per film)
        item_popularity = {
            movie_id: len(self.ratings[self.ratings['movieId'] == movie_id])
            for movie_id in self.ratings['movieId'].unique()
        }

        diversity = self.calculate_intra_list_diversity(recommendations)
        novelty = calculate_novelty(recommendations, item_popularity)

        return {
            "diversity": diversity,
            "novelty": novelty,
            "recommendations": recommendations
        }

    # --- Fungsi tambahan untuk Genre dan Enhanced Similarity ---

    def load_movie_genres(self, genres_path):
        """
        Memuat data genre film dan menyimpannya dalam bentuk yang mudah diakses.

        Parameters:
        - genres_path: Path ke file CSV yang berisi data film dan genre

        Returns:
        - Dictionary mapping movie_id ke list genre
        """
        df_movies = pd.read_csv('/content/drive/MyDrive/dataset/ml-latest-small/ml-latest-small/movies.csv')
        print("Sample data film:")
        print(df_movies.head())

        self.movie_genres = {}

        for _, row in df_movies.iterrows():
            movie_id = row['movieId']
            genres_str = row['genres']
            genres_list = genres_str.split('|')
            if genres_list == [''] or genres_list == ['(no genres listed)']:
                genres_list = []
            self.movie_genres[movie_id] = genres_list

        self.all_genres = set()
        for genres in self.movie_genres.values():
            self.all_genres.update(genres)

        print(f"Total film dengan data genre: {len(self.movie_genres)}")
        print(f"Total genre unik: {len(self.all_genres)}")

        return self.movie_genres

    def calculate_genre_similarity(self, movie_id1, movie_id2):
        """
        Menghitung similarity berdasarkan genre antara dua film.
        Menggunakan Jaccard similarity: intersection / union

        Parameters:
        - movie_id1: ID film pertama
        - movie_id2: ID film kedua

        Returns:
        - Float: Nilai similarity genre (0-1)
        """
        if not hasattr(self, 'movie_genres'):
            print("Data genre belum dimuat. Gunakan load_movie_genres() terlebih dahulu.")
            return 0.0

        genres1 = set(self.movie_genres.get(movie_id1, []))
        genres2 = set(self.movie_genres.get(movie_id2, []))

        if not genres1 or not genres2:
            return 0.0

        intersection = len(genres1.intersection(genres2))
        union = len(genres1.union(genres2))

        if union == 0:
            return 0.0

        return intersection / union

    def get_similar_movies_enhanced(self, movie_id, top_n=10, alpha=0.7):
        """
        Menemukan film yang mirip berdasarkan kombinasi latent factors dan kesamaan genre.

        Parameters:
        - movie_id: ID film yang menjadi referensi
        - top_n: Jumlah film yang akan dikembalikan
        - alpha: Bobot untuk similarity latent factor (1-alpha untuk similarity genre)

        Returns:
        - List of tuples: (movie_id, combined_similarity)
        """
        item_idx = self.get_item_idx(movie_id)

        if item_idx is None:
            print(f"Film dengan ID {movie_id} tidak ditemukan.")
            return []

        if not hasattr(self, 'movie_genres'):
            print("Data genre belum dimuat. Menggunakan similarity berbasis latent factor saja.")
            return self.get_similar_movies(movie_id, top_n)

        target_factors = self.item_factors[item_idx]
        similarities = []

        for idx, factors in enumerate(self.item_factors):
            if idx != item_idx:
                other_movie_id = self.get_original_movie_id(idx)
                latent_similarity = np.dot(target_factors, factors) / (np.linalg.norm(target_factors) * np.linalg.norm(factors))
                genre_similarity = self.calculate_genre_similarity(movie_id, other_movie_id)
                combined_similarity = alpha * latent_similarity + (1 - alpha) * genre_similarity
                similarities.append((other_movie_id, combined_similarity, latent_similarity, genre_similarity))

        sorted_similarities = sorted(similarities, key=lambda x: x[1], reverse=True)[:top_n]

        print(f"\nBreakdown similarity untuk film ID {movie_id}:")
        print(f"{'Movie ID':<10} {'Combined':<10} {'Latent':<10} {'Genre':<10}")
        print("-" * 40)
        for movie_id, combined, latent, genre in sorted_similarities[:5]:
            print(f"{movie_id:<10} {combined:.4f} {latent:.4f} {genre:.4f}")

        return [(movie_id, similarity) for movie_id, similarity, _, _ in sorted_similarities]

    def analyze_similarity_improvement(self, movie_id, top_n=10, alphas=[0.0, 0.3, 0.5, 0.7, 1.0]):
        """
        Menganalisis bagaimana perubahan bobot alpha mempengaruhi similarity.

        Parameters:
        - movie_id: ID film referensi
        - top_n: Jumlah rekomendasi yang akan dianalisis
        - alphas: List nilai alpha yang akan dicoba

        Returns:
        - Dictionary hasil analisis
        """
        results = {}

        print(f"Analisis peningkatan similarity untuk film ID {movie_id}")
        print("=" * 60)

        movie_title = "Unknown"
        if hasattr(self, 'movie_genres') and hasattr(self, 'movie_titles'):
            movie_title = self.movie_titles.get(movie_id, f"Film ID {movie_id}")

        print(f"Film referensi: {movie_title} (ID: {movie_id})")
        print("=" * 60)

        for alpha in alphas:
            print(f"\nDengan alpha = {alpha:.1f} (Latent: {alpha:.1f}, Genre: {1-alpha:.1f}):")
            similar_movies = self.get_similar_movies_enhanced(movie_id, top_n=top_n, alpha=alpha)
            results[alpha] = similar_movies

            print(f"Top {min(5, len(similar_movies))} film yang mirip:")
            for i, (similar_id, similarity) in enumerate(similar_movies[:5]):
                title = "Unknown"
                if hasattr(self, 'movie_titles'):
                    title = self.movie_titles.get(similar_id, f"Film ID {similar_id}")
                print(f"{i+1}. {title} (ID: {similar_id}) - Similarity: {similarity:.4f}")

        print("\nAnalisis perubahan peringkat berdasarkan alpha:")

        baseline_movies = [movie_id for movie_id, _ in results.get(1.0, [])]

        for alpha in [a for a in alphas if a != 1.0]:
            enhanced_movies = [movie_id for movie_id, _ in results.get(alpha, [])]
            new_entries = [m for m in enhanced_movies if m not in baseline_movies[:len(enhanced_movies)]]

            print(f"\nAlpha = {alpha:.1f}:")
            if new_entries:
                print(f"  Film baru yang muncul: {len(new_entries)}")
                for movie_id in new_entries[:3]:
                    title = "Unknown"
                    if hasattr(self, 'movie_titles'):
                        title = self.movie_titles.get(movie_id, f"Film ID {movie_id}")
                    print(f"  - {title} (ID: {movie_id})")
            else:
                print("  Tidak ada film baru yang muncul di top recommendations")

        print("\nRata-rata similarity berdasarkan alpha:")
        for alpha in alphas:
            avg_similarity = sum(sim for _, sim in results.get(alpha, []))
            if results.get(alpha):
                avg_similarity /= len(results.get(alpha))
                print(f"Alpha = {alpha:.1f}: {avg_similarity:.4f}")

        return results

    def load_movie_titles(self, movies_path):
        """
        Memuat judul film untuk meningkatkan interpretasi hasil.

        Parameters:
        - movies_path: Path ke file CSV yang berisi data film
        """
        df_movies = pd.read_csv(movies_path)
        self.movie_titles = {}
        for _, row in df_movies.iterrows():
            movie_id = row['movieId']
            title = row['title']
            self.movie_titles[movie_id] = title

        print(f"Berhasil memuat judul untuk {len(self.movie_titles)} film")
        return self.movie_titles

# Fungsi global untuk menghitung novelty (kebaruan)
def calculate_novelty(recommendations, item_popularity):
    """
    Menghitung novelty berdasarkan popularitas film.
    Novelty dihitung sebagai rata-rata self-information dari film-film yang direkomendasikan.
    """
    if not recommendations:
        return 0

    total_popularity = sum(item_popularity.values())
    novelty_sum = 0

    for item_id, _ in recommendations:
        probability = item_popularity.get(item_id, 0) / total_popularity
        if probability > 0:
            self_information = -np.log2(probability)
            novelty_sum += self_information

    average_novelty = novelty_sum / len(recommendations)
    return average_novelty

# Fungsi tambahan (tidak digunakan dalam demo)
def calculate_catalog_coverage(all_recommendations, total_items):
    unique_items = set()
    for user_recs in all_recommendations:
        for item_id, _ in user_recs:
            unique_items.add(item_id)
    coverage = len(unique_items) / total_items * 100
    return coverage

def calculate_category_diversity(recommendations, item_categories):
    category_counts = {}
    for item_id, _ in recommendations:
        if item_id in item_categories:
            for category in item_categories[item_id]:
                category_counts[category] = category_counts.get(category, 0) + 1
    total_categories = sum(category_counts.values())
    entropy = 0
    for count in category_counts.values():
        probability = count / total_categories
        entropy -= probability * np.log2(probability)
    max_entropy = np.log2(len(category_counts)) if category_counts else 0
    normalized_entropy = entropy / max_entropy if max_entropy > 0 else 0
    return normalized_entropy

# Fungsi utama untuk eksekusi model
def run_svd_recommender():
    # Split data menjadi data train dan test
    train_data, test_data = train_test_split(df, test_size=0.2, random_state=42)

    print(f"Jumlah data train: {len(train_data)}")
    print(f"Jumlah data test: {len(test_data)}")

    # Parameter model
    n_factors = 60
    learning_rate = 1e-2
    regularization = 0.1
    n_epochs = 70  # Bisa disesuaikan

    # Inisialisasi dan latih model
    svd = SVDMatrixFactorization(
        ratings=df,
        n_factors=n_factors,
        learning_rate=learning_rate,
        regularization=regularization,
        n_epochs=n_epochs
    )

    # Latih model
    train_rmse_log, test_rmse_log = svd.train(train_data, test_data)

    # Plot learning curve
    svd.plot_learning_curve()

    # Evaluasi model dengan RMSE
    print("\nEvaluasi Model:")
    print(f"Final Train RMSE: {train_rmse_log[-1]:.4f}")
    if test_rmse_log:
        print(f"Final Test RMSE: {test_rmse_log[-1]:.4f}")

    # Demo fitur-fitur
    print("\n===== Demo Fitur Model =====")

    # 1. Memprediksi rating yang hilang
    print("\n1. Memprediksi Rating yang Hilang:")
    sample_user = df['userId'].iloc[0]
    sample_movie = df['movieId'].iloc[0]
    predicted_rating = svd.predict(sample_user, sample_movie)
    print(f"Prediksi rating untuk User {sample_user} dan Film {sample_movie}: {predicted_rating:.2f}")

    # 2. Memprediksi film yang memiliki kemiripan (berdasarkan latent factors)
    print("\n2. Memprediksi Film yang Memiliki Kemiripan:")
    popular_movie = df['movieId'].value_counts().index[0]
    similar_movies = svd.get_similar_movies(popular_movie, top_n=5)
    print(f"Film yang mirip dengan Film ID {popular_movie}:")
    for movie_id, similarity in similar_movies:
        print(f"Film ID: {movie_id}, Similarity: {similarity:.4f}")

    # 3. Merekomendasikan film untuk pengguna
    print("\n3. Merekomendasikan Film untuk Pengguna:")
    active_user = df['userId'].value_counts().index[0]
    recommended_movies = svd.recommend_movies(active_user, top_n=5)
    print(f"Rekomendasi film untuk User {active_user}:")
    for movie_id, pred_rating in recommended_movies:
        print(f"Film ID: {movie_id}, Prediksi Rating: {pred_rating:.2f}")

    # Evaluasi kualitas rekomendasi untuk user aktif
    quality_metrics = svd.evaluate_recommendation_quality(active_user, num_recommendations=5)
    print("\n=== Metode Evaluasi Kualitas Rekomendasi ===")
    print("Diversity:", quality_metrics["diversity"])
    print("Novelty:", quality_metrics["novelty"])

    # 4. Demo fitur Enhanced Similarity dan Analisis Genre
    print("\n4. Analisis Kemiripan Genre dan Enhanced Similarity:")
    genres_path = '/content/drive/MyDrive/dataset/ml-latest-small/ml-latest-small/movies.csv'
    # Muat data genre dan judul film
    svd.load_movie_genres(genres_path)
    svd.load_movie_titles(genres_path)
    # Pilih film referensi untuk analisis (misalnya film pertama dari data genre)
    reference_movie = list(svd.movie_genres.keys())[0]
    enhanced_similar_movies = svd.get_similar_movies_enhanced(reference_movie, top_n=5, alpha=0.7)
    print(f"\nFilm yang mirip dengan Film ID {reference_movie} berdasarkan enhanced similarity:")
    for movie_id, similarity in enhanced_similar_movies:
        title = svd.movie_titles.get(movie_id, f"Film ID {movie_id}")
        print(f"Film: {title} (ID: {movie_id}) - Combined Similarity: {similarity:.4f}")

    # Analisis perubahan similarity berdasarkan berbagai nilai alpha
    svd.analyze_similarity_improvement(reference_movie, top_n=5, alphas=[0.0, 0.3, 0.5, 0.7, 1.0])

    return svd

# Jalankan model
if __name__ == "__main__":
    svd_model = run_svd_recommender()
```
